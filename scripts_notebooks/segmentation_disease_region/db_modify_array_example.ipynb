This notebook illustrates 'modify_array' function, using image and mask examples. This function is called when creating a database.

```python
import nibabel as nib
from db_modify_array import * 
import matplotlib.pyplot as plt
```

```python
## Load data, retrieve numpy arrays
image = nib.load('example_image.nii.gz')
image = image.get_fdata()

mask = nib.load('example_mask.nii.gz')
mask = mask.get_fdata()

## Modify arrays
modified_image = modify_array(image)
modified_mask = modify_array(mask)
```

```python
## Output arrays dimension and data type
print('Original arrays size and data type:', image.shape, mask.shape,
                                             image.dtype, mask.dtype)

print('Modified arrays size and data type:', modified_image.shape, modified_mask.shape,
                                             modified_image.dtype, modified_mask.dtype)

## Visualize image and mask slcies
## Choose slice
s = 12

plt.figure(figsize = (10,8))

plt.subplot(221)
plt.imshow(image[:,:,s], cmap = 'gray')
plt.title('Original image slice', fontsize = 14)
plt.axis('off')
plt.colorbar()

plt.subplot(222)
plt.imshow(modified_image[s,0,:,:], cmap = 'gray')
plt.title('Modified image slice', fontsize = 14)
plt.axis('off')
plt.colorbar()

plt.subplot(223)
plt.imshow(mask[:,:,s], cmap = 'gray')
plt.title('Original mask slice', fontsize = 14)
plt.axis('off')
plt.colorbar(ticks = [0,1])

plt.subplot(224)
plt.imshow(modified_mask[s,0,:,:], cmap = 'gray')
plt.title('Modified mask slice', fontsize = 14)
plt.axis('off')
plt.colorbar(ticks = [0,1])

plt.show()
```
